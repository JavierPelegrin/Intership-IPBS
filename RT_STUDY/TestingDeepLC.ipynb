```python
class pep():
    def __init__(self,peptide_id,seq,elutionTime):
        self.peptide_id = peptide_id
        self.seq = seq
        self.elutionTime = elutionTime


def sliding_window(elements, window_size):    
    if len(elements) <= window_size:
       return elements
    for i in range(len(elements)- window_size + 1):
        yield elements[i:i+window_size]


entities = []
for attribute0, attribute1, attribute2 in zip(df_pred["peptide_id"].values,df_pred["seq"].values, df_pred['tr'].values):
    entity = pep(attribute0 ,attribute1, attribute2)
    entities.append(entity)

print("all")

slidingSize = 400
half_sliding_size = slidingSize // 2

#sliding_window = sliding_window(sorted(entities, key=lambda x: x.elutionTime), slidingSize)
sorted_entities = sorted(entities, key=lambda x: x.elutionTime)
n_entities = len(sorted_entities)

data = []
print("all")
i = half_sliding_size

while i < n_entities - half_sliding_size:
    
    current_ion = sorted_entities[i]

    j = 0
    while j < slidingSize and i + j < n_entities:

        #current_ion = co_eluting_master_entities[half_sliding_size]
        #for pepB in co_eluting_master_entities:

        pepB = sorted_entities[i + j]
        if pepB != current_ion and len(current_ion.seq) == len(current_ion.seq):
            data.append([current_ion.seq,pepB.seq, current_ion.peptide_id, pepB.peptide_id,current_ion.elutionTime,pepB.elutionTime,pepB.elutionTime-current_ion.elutionTime])

        j += 1

    i += 1

df_pred = pd.DataFrame(data, columns=['seqA', 'seqB', 'pepIdA', "pepIdB",'elutionTimeA', 'elutionTimeB', 'deltaElutionTime'])
```

```python

def pass_to_EncyclopeDIA_style(seq,mod):
    DIA = {
        "Carbamidomethyl" : ( "+57.021464", ("C")),
        "Oxidation": ("+15.99491" ,("M")),
        "Phosphorylation": ("+79.966331" ,("S", "T", "Y")),
        "Acetyl": ("+42.010565" ,("K", "N-term")),
        "Succinylation": ("+101.023869" ,("K")),
        "Ubiquitylation": ("+114.042927" ,("K")),
        "Mono-methylation": ("+14.015650" ,("K", "R")),
        "Di-methylation": ("+28.031300" ,("K", "R")),
        "Tri-methylation": ("+42.046950" ,("K")),
        "TMT0": ("+224.152478" ,("K", "N-term")),
        "TMT10": ("+229.162932" ,("K", "N-term")),
        "Pyroglutamate": ("-18.010565" ,("N-term E")),
        "Pyroglutamate": ("-17.026549" ,("N-term","Q")),
        "Cyclized S-CAM-Cys": ("+39.994915" ,("N-term", "C"))
    }
    mod = mod.split("|")
    
    new_seq = ""
    while len(mod) > 0:
        n = int(mod.pop(0))
        new_seq = f"{new_seq}{seq[:n]}[{DIA[mod.pop(0)][0]}]"
        seq = seq[n:]
    return f"{new_seq}{seq}"
    
```
